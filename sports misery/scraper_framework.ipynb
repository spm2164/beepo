In the real version, I'll be reusing this and generating the url 1 year at a time, and then again for each other sport. Hopefully the other objects are similar, but everything else should adapt to the other sites well.

```python
csv_button=driver.find_element_by_xpath("//*[@id='all_expanded_standings_overall']/div[1]/div/span[5]")
csv_button.click()
```

The key in all of the (sports)-reference.com pages is that it gives you a table that is kind of useless, but there is a (not) button that you can click and it just spits out a csv. Until you click that (not) button, which is actually a span class, whatever that is, the csv will not show up in the page source. So, we click by xpath reference, then grab the source and strip out the csv.

```python
doc=BeautifulSoup(driver.page_source, 'html.parser')
```

```python
csv_table=doc.find(id='csv_expanded_standings_overall').text
```

This is finally my actual table. Since this is already in csv, I don't need to do anything else to export it.

```python
csv_table
```

```python
csv_file = open("Output.csv", "w")
csv_file.write(csv_table)
csv_file.close()
```
