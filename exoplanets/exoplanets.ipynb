```python
fig, ax = plt.subplots(figsize=(20,20))
plt.xlim(0,1)
plt.ylim(0,1)
ax.set_axis_bgcolor('black')
scale=1000
plt.scatter(0,0,s=scale*9.95**2,c='yellow',linewidths=0.1)

r=pi/4
rad=sqrt(9.95**2*pi/25000)
sun_x=rad*cos(r)
sun_y=rad*sin(r)

for index, planet in df_ssp.iterrows():    
    r=pi/4
    x=planet['Mean Distance from Sun (AU)']*cos(r)+sun_x
    y=planet['Mean Distance from Sun (AU)']*sin(r)+sun_y
    plt.scatter(x, y,
                s=scale*planet['radj']**2, c='green',linewidths=0.1)
    plt.annotate(planet['Name'], xy=(x+0.002,y+0.002), color='white')
    
for index, planet in df_ro.iterrows():    
    r=pi/2*random.random()
    rad=sqrt(9.95**2*pi/25000)
    sun_x=rad*cos(r)
    sun_y=rad*sin(r)
    x=planet['pl_orbsmax']*cos(r)+sun_x
    y=planet['pl_orbsmax']*sin(r)+sun_y
    plt.scatter(x, y, s=scale*planet['pl_radj']**2, c='white',linewidths=0.2,alpha=0.5)

plt.tight_layout()
plt.savefig('planets_another_attempt.pdf')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.xlim(-0.5,0.5)
plt.ylim(-0.5,0.5)
ax.set_axis_bgcolor('black')
plt.scatter(0,0,s=9.95,c='yellow',linewidths=0.1)

for index, planet in df_ssp.iterrows():    
    r=pi/4
    x=planet['Mean Distance from Sun (AU)']*cos(r)
    y=planet['Mean Distance from Sun (AU)']*sin(r)
    plt.scatter(x, y,
                s=planet['radj'], c='green',linewidths=0.1)
    plt.annotate(planet['Name'], xy=(x+0.002,y+0.002), color='white')

for index, planet in df_ro.iterrows():    
    r=2*pi*random.random()
    x=planet['pl_orbsmax']*cos(r)
    y=planet['pl_orbsmax']*sin(r)
    plt.scatter(x, y, s=planet['pl_radj'], c='white',linewidths=0.2,alpha=1)
    
plt.savefig('planets_no_scaling_merc_zoom.pdf')
plt.show()
```
